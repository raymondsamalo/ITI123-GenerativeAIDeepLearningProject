```python
import sys
sys.path.append('.')
# from yolo import ODIRYOLOTrainer
from yolo_trainer import ODIRYOLOTrainer
```

```python
run_name = "yolo11n-cls-preprocessed"
model_version = 'yolo11n-cls'
data_dir = "./ODIR-2019/YOLO/preprocessed"
project_name = "odir-2019-yolo"
```

## Train

```python
import wandb

wandb.init(project=project_name, name=run_name)
trainer = ODIRYOLOTrainer(model_version=model_version, use_wandb=True,wandb_run=wandb.run)
trainer.train_from_folder(data_folder=data_dir, epochs=30, batch_size=-1, imgsz=512, patience=15,run_name=run_name)
```

## Validate

```python
trainer.evaluate_on_validation_set(data_folder=data_dir, batch_size=-1, imgsz=512)
```

## Test

```python
trainer.evaluate_on_test_set(data_folder=data_dir, batch_size=-1, imgsz=512)
```

## Save

```python
# 4. Save trained model
trainer.save_model(
    save_dir='./ODIR-2019/results',
    model_name=run_name,
    save_format='pt'
)
```
