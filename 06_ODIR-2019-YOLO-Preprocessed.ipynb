```python
import sys
sys.path.append('.')
from yolo import ODIRYOLOTrainer
```

```python
run_name = "yolo-preprocessed"
model_version = 'yolo11n-cls'
data_dir = "./ODIR-2019/YOLO/preprocessed"
project_name = "odir-2019-yolo"
```

## Train

```python
import wandb

wandb.init(project=project_name, name=run_name)
trainer = ODIRYOLOTrainer(model_version=model_version, run_name=run_name)
data_yaml=trainer.create_data_yaml(data_dir=data_dir)
trainer.train(data_yaml=data_yaml, epochs=100, batch_size=16, img_size=512, patience=30)
```

## Validate

```python
trainer.evaluate_model(data_yaml, split='val', imgsz=512)
```

## Test

```python
trainer.evaluate_model(data_yaml, split='test', imgsz=512)
```

## Save

```python
# 4. Save trained model
trainer.save_model(
    save_dir='./ODIR-2019/results',
    model_name=run_name,
    save_format='pt'
)
```
